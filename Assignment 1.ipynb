````r
---
title: "Assignment #1"
author: "Anthony J. Kulowski"
date: "September 17, 2018"
output:
  word_document: default
  pdf_document: default
---

```{r setup, include=FALSE}
knitr::opts_chunk$set(echo = TRUE)
```


```{r, echo=FALSE}
titanic=read.csv("C:/Users/student/Documents/Bryant University/4 - Senior Year 2018-19/2018 FA/MATH 421/titanic.csv")

library(ggplot2)
titanic$Survived = factor(titanic$Survived)

```


#Graph #1
```{r, echo=FALSE}
ggplot(titanic) + geom_histogram(binwidth = 3, mapping = aes(x = Age, fill = Sex), position = "dodge")
```
This histogram shows the distribution of age for both males and females. We can see that the male to female is relatively even for 0 to 18 then the count of males takes the majority by a large amount. We can also see that there weren't any females above 61.



#Graph #2
```{r, echo=FALSE}
ggplot(titanic) + geom_bar(aes(Sex))
```
While similar to graph 1, this barchart provides a high level view of the amount of males and females onboard. We can see that there were nearly twice as many males as females. Given the year being in 1912, we would expect to see a vast difference in count.

#Graph #3
```{r, echo=FALSE}
ggplot(titanic) + geom_boxplot(mapping = aes(y = Age))
```
This boxplot shows us that the majority of occupants were in between the age of 20 to 40 with a few outliers with a higher age.

#Graph #4
```{r, echo=FALSE}
ggplot(titanic) + geom_bar(mapping = aes(x = Embarked, fill = Sex), position = "dodge")
```
This bar chat shows us the amount of male and females from the point they embarked from. We can see that the ratio of male to female is pretty even for both Cherbourg and Queenstown. We see a major difference in the Southhampton port.

#Graph #5
```{r, echo=FALSE}
ggplot(titanic) + geom_bar(mapping = aes(x = Pclass, fill = Sex),position = "dodge")
```
This barchart contrasts the class males and females were in. For first and second class was mainly an even splot but there was a big drop off for females when it reaches third class. This is something we could expect because the men traveling in third class could be mainly due to work reasons where as women wouldn't be as likely to travel third class at the time.

#Graph #6
```{r, echo=FALSE}
ggplot(titanic) + geom_dotplot(binwidth = 1.5, mapping = aes(x = Age, fill = Survived), position = "dodge")
```
This dot plot compares the people by age who survived. The younger group of less than 18 so those considered children have the highest likelihood of surving. We see only about 1 for every 3 people above that age survived.

#Graph #7
```{r, echo=FALSE}
ggplot(data = titanic) + 
  geom_smooth(mapping = aes(x = Fare, y = Age, linetype = factor(Survived)))
```
This smoothed conditional means graph identifies the people who paid more in fare for there tickets were more likely to survive the sinking.
````
